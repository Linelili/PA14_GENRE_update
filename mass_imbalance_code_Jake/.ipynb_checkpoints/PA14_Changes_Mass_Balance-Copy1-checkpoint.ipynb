# rxn13808 Results 

```python
#reaction name and model creation

rxn_name = 'rxn13808'
rxn_model = copy.deepcopy(model)
rxn_name_list = [rxn_name]

#Reaction as Written

'''
PROBLEMS: Charge imbalance +2 on the right right side 

3.0 cpd00004_c + 4.0 cpd00067_c + cpd00075_c <=> cpd00001_c + 3.0 cpd00003_c + cpd00985_c

3 NADH [0] + 4 H+ [0] + 1 Nitrite [0] -> 1 H2O [0] + 3 NAD [0] + 1 NH4OH [0]

NH4OH has charge 2 in modelseed and in model
Nh3 H2o is NH4OH... Recommended changing to neutral charge 

'''

#Reaction Changes in the cahnged reaction model

rxn_model.metabolites.get_by_id('cpd00985_c').charge = 0 

#Quick Mass balance results

print('Mass Balance Results:', rxn_model.reactions.get_by_id(rxn_name).check_mass_balance())

#Reaction Change Results on Objective Values

rxn_result = function_farm.minimal_media_varied_carbon(rxn_model)
rxn_Objectives = rxn_result[0]
rxn_comparison_results = function_farm.compared(Names_Objectives_Unchanged, rxn_Objectives) # compare the results

#check for free mass

# function_farm.checkfreemass(rxn_model)

#reaction results on fluxes and gene deletions

gene_flux_result_rxn = function_farm.results_genes_flux(rxn_model, rxn_name_list, [1,1])
rxn_name_series = pd.Series(rxn_name_list)
gene_flux_result_rxn[1]['Reaction_Name'] = rxn_name_series.values
gene_flux_result_rxn[1].to_excel(rxn_name + '_flux.xlsx', sheet_name='sheet1', index=False)
print('\nReaction File created called:',rxn_name +'_flux.xlsx')
```

# rJB00234 Results 

```python
#reaction name and model creation
rxn_name = 'rJB00234'
rxn_model = copy.deepcopy(model)
rxn_name_list = [rxn_name]

#Reaction as Written

'''
Problem: Charge imbalance -1 

Solution: cpd00489_c is 0. Should be -1.

Rxn Equation: cpd00067_e + cpd00489_e <=> cpd00067_c + cpd00489_c
Rxn Equation Change: NONE

'''

#Reaction Changes in the cahnged reaction model

rxn_model.metabolites.get_by_id('cpd00489_e').charge = -1

#Quick Mass balance results

print('Mass Balance Results:', rxn_model.reactions.get_by_id(rxn_name).check_mass_balance())

#Reaction Change Results on Objective Values

rxn_result = function_farm.minimal_media_varied_carbon(rxn_model)
rxn_Objectives = rxn_result[0]
rxn_comparison_results = function_farm.compared(Names_Objectives_Unchanged, rxn_Objectives) # compare the results

#reaction results on fluxes and gene deletions

gene_flux_result_rxn = function_farm.results_genes_flux(rxn_model, rxn_name_list, [1,1])
rxn_name_series = pd.Series(rxn_name_list)
gene_flux_result_rxn[1]['Reaction_Name'] = rxn_name_series.values
gene_flux_result_rxn[1].to_excel(rxn_name + '_flux.xlsx', sheet_name='sheet1', index=False)
print('\nReaction File created called:',rxn_name +'_flux.xlsx')

```
