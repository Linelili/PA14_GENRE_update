```python
def changeMedia_PA_LJD(model, media):
     LB_open_exchanges = ['EX_cpd00001(e)', 
                     'EX_cpd00009(e)',
                     'EX_cpd00011(e)',
                     'EX_cpd00021(e)',
                     'EX_cpd00034(e)',
                     'EX_cpd00048(e)',
                     'EX_cpd00058(e)',
                     'EX_cpd00205(e)',
                     'EX_cpd00254(e)',
                     'EX_cpd00971(e)',
                     'EX_cpd01012(e)',
                     'EX_cpd00067(e)']

    LB_limited_exchanges = ['EX_cpd00023(e)',
                            'EX_cpd00027(e)',
                            'EX_cpd00033(e)',
                            'EX_cpd00035(e)',
                            'EX_cpd00039(e)',
                            'EX_cpd00041(e)',
                            'EX_cpd00051(e)',    
                            'EX_cpd00054(e)',    
                            'EX_cpd00060(e)',    
                            'EX_cpd00065(e)',    
                            'EX_cpd00066(e)',    
                            'EX_cpd00069(e)',    
                            'EX_cpd00084(e)',    
                            'EX_cpd00107(e)',    
                            'EX_cpd00119(e)',    
                            'EX_cpd00129(e)',    
                            'EX_cpd00156(e)',    
                            'EX_cpd00161(e)',    
                            'EX_cpd00305(e)',   
                            'EX_cpd00322(e)',    
                            'EX_cpd00092(e)',    
                            'EX_cpd00307(e)',
                            'EX_cpd03091(e)']

    for i in range(0,len(model.exchanges)):
        # Set all EX rxns lb = 0, up = 1000
        if model.exchanges[i].name[0:2] == 'EX':
            model.exchanges[i].upper_bound = 1000
            model.exchanges[i].lower_bound = 0
        # Set aerobic exchange (O2) to 20 mmol/gDW/hr 
        if model.exchanges[i].id[0:11] == 'EX_cpd00007':
            model.exchanges[i].lower_bound = -20
    
        # Media == 1 Change bounds to LB
        if media == 1:
            # Set open lb exchanges to -1000
            for j in range(0,len(LB_open_exchanges)):
                if model.exchanges[i].id[0:11] == LB_open_exchanges[j][0:-3]:
                    model.exchanges[i].lower_bound = -1000
            # Set limited lb exchanges to -10
            for k in range(0,len(LB_limited_exchanges)):
                if model.exchanges[i].id[0:11] == LB_limited_exchanges[k][0:-3]:
                    model.exchanges[i].lower_bound = -10
        else:
            print('unrecognized media condition. Please enter 1 for LB; 2 for SCFM; 3 for minimal media')
    return(model) 
```
